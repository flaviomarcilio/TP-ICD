```python
# Talvez retirar as colunas que possuem valores nulos seja menos radical, pois talvez sejam informações menos relevantes
# camp_bras_full = camp_bras_full.dropna()
# camp_bras_full.info()
# camp_bras_full
```

```python
base_full = camp_bras_full.drop(columns=['formacao_mandante', 'formacao_visitante', 'tecnico_mandante', 'tecnico_visitante'])
base_full.info()
```

```python
base_full.rename(columns={'ID': 'partida_id', 'rodata': 'rodada'}, inplace=True)
base_full.info()
```

```python
base_full['horario_partida'] = pd.to_datetime(base_full['data'] + ' ' + base_full['hora'])
```

```python
base_full = base_full.drop(columns=['data', 'hora'])
base_full.info()
```

```python
# Quantidade de Partidas por ano
base_full.groupby(base_full['horario_partida'].dt.year)['partida_id'].count()
```
